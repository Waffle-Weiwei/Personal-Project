```python
import pandas as pd
girls12_df = pd.read_csv("/content/drive/MyDrive/Data Science/Heights_Girls12_Cleaned.csv")
```

```python
girls12_df.head()
```

```python
from scipy.stats import linregress
result = linregress(girls12_df['Year'],girls12_df['Mean height'])
result
```

Slope: 0.0835

p-value: 1.3643 e-20

There is statistically significant evidence of a relationship between year and mean height for 12-year-old girls from 1985 to 2019, so we reject the null hypothesis of no change.

```python
import numpy as np
boot_slopes = []
for _ in range(5000):
  sample = girls12_df.sample(n=len(girls12_df), replace=True)
  boot_slopes.append(linregress(sample['Year'], sample['Mean height']).slope)
np.percentile(boot_slopes, [2.5, 97.5])
```

The estimated yearly increase is plausibly between about 0.0727 and 0.0942 cm per year, and that interval doesn't include 0, which points to an upward trend.

- confounder: average childhood nutrition quality over time
 - subgroup check: early years vs recent years

```python
early = girls12_df[girls12_df['Year'] < 2002]
recent = girls12_df[girls12_df['Year'] >= 2002]
early.shape, recent.shape
early['Mean height'].mean(), recent['Mean height'].mean()
```
